# Тести для малих вибірок та пропорцій



**Завдання 1**. E-commerce компанія після редизайну сайту підозрює, що **середній час до покупки (time-to-purchase)** користувачів **збільшився**.
Історично середній час від першого заходу на сайт до покупки становив $\mu_0 = 123.7$ хвилин.

Команда аналітиків випадково обрала дані **7 покупців після редизайну**:
`128, 135, 121, 142, 126, 151, 123`

З допомогою наявних даних зʼясуйте, чи збільшився середній час покупки після редизайну.

Для цього виконайте наступні 6 кроків. Правильне виконання кожного з кроків оцінюється в 1 бал.

1. Запишіть нульову та альтернативну гіпотези і визначте тип тесту.
2. Обчисліть вибіркові статистики: $\bar x$, $s$, $n$.
3. Оберіть тип тесту та виконайте його (знайдіть критичне значення тесту, статистику тесту та р-значення) будь-яким пасуючим способом, щоб перевірити гіпотезу на рівні значущості $\alpha = 0.10$.
4. Прийміть рішення, чи відхиляєте ви гіпотезу $H_0$ використовуючи p-value.
5. Напишіть висновок: чи справді редизайн сайту подовжив час до покупки?
6. Чи зміниться ваше рішення при зміні рівня значущості на $\alpha = 0.05$.


```python
import numpy as np
from scipy import stats

av_time = np.array([128, 135, 121, 142, 126, 151, 123])
mu0 = 123.7
alpha = 0.1
alpha1 = 0.05

x_bar = np.mean(av_time)
s = np.std(av_time, ddof=1)
n = len(av_time)

t_stat = (x_bar - mu0) / (s / np.sqrt(n))
df = n - 1
t_crit = stats.t.ppf(1 - alpha, df)
t_crit1 = stats.t.ppf(1 - alpha1, df)
p_value = 1 - stats.t.cdf(t_stat, df)
```

```python
print("-" * 40)
print("РОЗВ'ЯЗОК #1: гіпотеза")
print("-" * 40)

print(f"H₀: μ = {mu0} (середній час покупки)")
print(f"H₁: μ > {mu0} (чи збільшився середній час покупки після редизайну)")
print(f"Тип тесту: правосторонній (one-tailed)")
print()

print("-" * 40)
print("РОЗВ'ЯЗОК #2: вибіркова статистика")
print("-" * 40)

print(f"Вибіркове середнє x̄: {x_bar:.2f}")
print(f"Вибіркове стандартне відхилення s: {s:.2f}")
print(f"Кількість спостережень n: {n}")
print()

print("-" * 40)
print("РОЗВ'ЯЗОК #3: t-статистика")
print("-" * 40)

print(f"t-статистика: {t_stat:.4f}")
print(f"Критичне значення (df={df}): {t_crit:.4f}")
print(f"p-value: {p_value:.4f}")
print()

print("-" * 40)
print("РОЗВ'ЯЗОК #4: Висновок (використовуючи p-value)")
print("-" * 40)

if p_value < alpha:
    print(f"p_value:{p_value:.4f} < alpha:{alpha}")
    print(f"Відхиляємо H₀: середній час покупки після редизайну збільшився на рівні значущості {alpha}")
else:
    print(f"{p_value:.4f} > {alpha}")
    print("Не відхиляємо H₀: доказів недостатньо")
print()

print("-" * 40)
print("РОЗВ'ЯЗОК #5: Висновок(α=0.10)")
print("-" * 40)

if t_stat > t_crit:
    print(f"t_stat:{t_stat:.4f} > t_crit:{t_crit:.4f}")
    print(f"Відхиляємо H₀: середній час покупки після редизайну збільшився на рівні значущості {alpha}")
else:
    print(f"t_stat:{t_stat:.4f} < t_crit:{t_crit:.4f}")
    print("Не відхиляємо H₀: доказів недостатньо")
print()

print("-" * 40)
print("РОЗВ'ЯЗОК #6: Висновок (α=0.05)")
print("-" * 40)

if t_stat > t_crit1:
    print(f"t_stat:{t_stat:.4f} > t_crit:{t_crit1:.4f}")
    print(f"Відхиляємо H₀: середній час покупки після редизайну збільшився на рівні значущості {alpha1}")
else:
    print(f"t_stat:{t_stat:.4f} < t_crit:{t_crit1:.4f}")
    print("Не відхиляємо H₀: доказів недостатньо")


```
