````python
from math import pi


def calc_area(radius):
    """Обчислює площу кола за радіусом."""
    return pi * radius ** 2


def calc_circumference(radius):
    """Обчислює довжину кола за радіусом."""
    return 2 * pi * radius


# Ініціалізація змінної
radius = 5

# Виведення результатів
print(
    "Area:", calc_area(radius),
    "Circumference:", calc_circumference(radius)
)
````

1.   Замінено from math import * на конкретний імпорт (pi)

8.   Використано зрозумілі назви параметрів (radius замість r)

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6

df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

# --- ПАМ’ЯТЬ ДО ---
memory_before = df.memory_usage(deep=True).sum()

# --- ОПТИМІЗАЦІЯ (перетворення в category) ---
df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

# --- ПАМ’ЯТЬ ПІСЛЯ ---
memory_after = df.memory_usage(deep=True).sum()

# --- РЕЗУЛЬТАТ ---
print(f"Пам'ять до: {memory_before / (1024**2):.2f} MB")
print(f"Пам'ять після: {memory_after / (1024**2):.2f} MB")
print(f"Зменшення у разів: {memory_before / memory_after:.2f}")
````

````python
def calculate_variance(numbers):
    """Обчислює дисперсію списку чисел."""

    mean = sum(numbers) / len(numbers)

    mean_squares = sum(n ** 2 for n in numbers) / len(numbers)

    variance = mean_squares - mean ** 2

    return variance


input_list = [75, 80, 85, 90, 95]

result = calculate_variance(input_list)
print(result)
````
